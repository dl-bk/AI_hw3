# Завдання 6
Перевірте чи правильно розраховано загальну вартість покупки. Виведіть ті рядки де є помилка. Виправте дані в таблиці

```python
incorrect_amt = df.loc[df['units'] * df['unit_price'] != df['sale_amt']]
incorrect_amt_idxs = incorrect_amt.index
for idx in incorrect_amt_idxs:
  print("Problem row:")
  print(df.iloc[[idx]])
  df.loc[idx, 'sale_amt'] = df.loc[idx, 'units'] * df.loc[idx, 'unit_price']
  print("data corrected")




```
